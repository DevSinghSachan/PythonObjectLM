# Covariance over a tree

Get the covariance by using the Laplacian

```python
%load_ext autoreload
%autoreload 2
%config InlineBackend.figure_format = 'svg'
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as colormap
from IPython.display import Image
from pygraphviz import AGraph
import numpy as np

DEFAULT_COLORMAP = colormap.RdYlGn
    
def norm_pdf(y, cov, inv = True):
    if inv:
        return np.exp(-0.5 * np.dot(np.dot(y.T, np.linalg.inv(cov)),y))
    else:
        return np.exp(-0.5 * np.dot(np.dot(y.T, cov),y))
    
def entropy(p_x):
    """
    Entropy:
    H(p_x) = - sum p_x * log(p_x)
    
    Inputs
    ------
    
    p_x ndarray : the probability distribution
    
    Outputs
    -------
    
    entropy float : the entropy of the distribution
    """
    return (-p_x * np.log(p_x)).sum()
    
    
def to_hex(rgba):
    r, g, b, a = rgba
    return '#%02x%02x%02x' % (int(r * 255), int(g * 255), int(b * 255))

example_graph = {
    "World": {
    "News": {
        "World news": {},
        "Tech news": {}
    },
    "Science" : {
        "solar": {},
        "CS":{}
    },
    "Religion" : {
        "Temples":{},
        "Study":{}
    },
    "Politics": {
        "USA Politics": {},
        "French Politics": {}
    }
    }}

class DrawableGraph(AGraph):
    def render(self, path, prog = "dot"):
        super().draw(path, prog = prog)
        return Image(path)
    
    def node(self, *args, **kwargs):
        return self.add_node(*args, **kwargs)
    
    def edge(self, *args, **kwargs):
        return self.add_edge(*args, **kwargs)
    
    def _update_graph(self, dict_node, code = [], name = None):
        if name == None and len(dict_node.keys()) > 1:
            name = 'ROOT'
        elif name == None and len(dict_node.keys()) == 1:
            return self._update_graph(dict_node[list(dict_node.keys())[0]], code, list(dict_node.keys())[0])
        
        self.node(code, label = "%s %r" % (name, code))
        self.name2code[name] = code
        # connect to parents
        if len(code) > 0:
            self.edge(code[:-1], code)

        for i, subkey in enumerate(dict_node.keys()):
            self._update_graph(dict_node[subkey], code + [i], subkey)
        return self
    
    def __init__(self, dictionary = None, cmap = None, **kwargs):
        super().__init__(**kwargs)
        self.name2code = {}
        self.edge_attr["arrowType"] = "none"
        cmap = cmap if cmap is not None else DEFAULT_COLORMAP
        self.cmap = colormap.ScalarMappable(norm = lambda x: x, cmap = cmap)
        if dictionary is not None:
            self._update_graph(dictionary)
            
    def prob2color(self, value):
        return to_hex(self.cmap.to_rgba(value))
    
    def get_node_by_name(self, name):
        return self.get_node(self.name2code[name])
            
    def apply_conditional_probabilities(self,
                                          labels,
                                          probs):
        for i, label in enumerate(labels):
            node = self.get_node_by_name(label)
            node.attr["fillcolor"] = self.prob2color(probs[i,0])
            node.attr["style"] = "filled"
            
        return self
    
    def apply_observations(self,
                               positive = [],
                               negative = []):
        for label in positive:
            node = self.get_node_by_name(label)
            node.attr["color"] = self.prob2color(1.0)
            node.attr["fillcolor"] = "transparent"
            node.attr["style"] = "filled"
            node.attr["shape"] = "rect"
            
        for label in negative:
            node = self.get_node_by_name(label)
            node.attr["color"] = self.prob2color(0.0)
            node.attr["fillcolor"] = "transparent"
            node.attr["style"] = "filled"
            node.attr["shape"] = "rect"
        return self

class DictTree:
    def __init__(self, graph, edge_weight = 1.):
        self.edge_weight = edge_weight
        index2label, label2index, connections = self.convert_dict_to_matrix_labels(graph)
        self.index2label = index2label
        self.label2index = label2index
        self._underlying_graph = graph
        self.create_matrices(connections, len(index2label))
        
    @classmethod
    def convert_dict_to_matrix_labels(cls, graph,
                                  entry_key = None,
                                  index2label = None,
                                  label2index = None,
                                  connections = None):
        
        index2label = [] if index2label is None else index2label
        label2index = {} if label2index is None else label2index
        connections = {} if connections is None else connections
    
        # create a root key if need be
        if entry_key == None and len(graph.keys()) == 1:
            entry_key = list(graph.keys())[0]
            graph = graph[entry_key]
        elif entry_key == None:
            entry_key = "ROOT"

        if label2index.get(entry_key, None) is not None:
            entry_key_index = label2index[entry_key]
        else:
            entry_key_index = len(label2index)
            label2index[entry_key] = entry_key_index
            index2label.append(entry_key)

        for subkey in graph.keys():
            # create child
            index2label.append(subkey)
            index = len(label2index)
            label2index[subkey] = index
            # connect parent to child:
            if connections.get(entry_key_index, None) is not None:
                connections[entry_key_index].append(index)
            else:
                connections[entry_key_index] = [index]

            # recurse:
            cls.convert_dict_to_matrix_labels(graph[subkey],
                                   entry_key = subkey,
                                   connections = connections,
                                   index2label = index2label,
                                   label2index = label2index)
        return (index2label, label2index, connections)
        
    def create_matrices(self, connections, n):
        self._adjacency_matrix = np.zeros([n, n], dtype=np.int32)
        self._degree_matrix    = np.zeros(n, dtype=np.int32)
        for key, connected_to in connections.items():
            self._adjacency_matrix[key, connected_to] = 1
            self._adjacency_matrix[connected_to, key] = 1
            self._degree_matrix[key] += len(connected_to)
            self._degree_matrix[connected_to] += 1
            
    def laplacian(self):
        return np.diag(self.degree_matrix) - self.adjacency_matrix
    
    @staticmethod
    def monte_carlo_covariance(degree_matrix,
                               adjancency_matrix,
                               sigma = None,
                               num_steps = 50,
                               num_trials = 100):
        variance = sigma**2 if sigma is not None else None
        # for each node we run several trials
        end_pos = np.zeros_like(adjancency_matrix)
        for node in range(degree_matrix.shape[0]):
            for trial in range(num_trials):
                current_position = node
                for step in range(num_steps):
                    # choose random direction:
                    if variance is not None and np.random.rand() < variance:
                        continue
                    direction = np.random.randint(degree_matrix[current_position])
                    # find prob of transition:
                    next_node = adjancency_matrix[current_position, :].nonzero()[0][direction]
                    if np.random.rand() < adjancency_matrix[current_position, next_node]:
                        current_position = next_node
                end_pos[node, current_position] += 1.
                end_pos[current_position, node] += 1.

        end_pos /= num_trials

        return end_pos
    
    def approximate_covariance(self, sigma = None,
                               num_steps = 50,
                               num_trials = 100):
        """
        Get a Monte Carlo estimate of the covariance by
        performing random walks across it (akin to finding
        the infinite power of this matrix), complexity is O(n * num_ste
        
        Inputs
        ------
        
         num_steps: number of steps each random walk takes
        num_trials: number of samples taken for each dimension
                    of matrix
        
        """
        return self.monte_carlo_covariance(
            self._degree_matrix,
            self.adjacency_matrix,
            sigma = sigma,
            num_steps = num_steps,
            num_trials = num_trials)
    
    def covariance(self, sigma = 5):
        """
        Get the covariance of this graph by taking the inverse
        of the sum of a laplacian and diagonal matrix representing
        a non correlated multivariate gaussian:
        
            K = (laplacian + 1/sigma^2 * I)^{-1}
        
        """
        if sigma is None:
            return np.linalg.inv(self.laplacian())
        else:
            return np.linalg.inv(self.laplacian() + (np.eye(len(self.index2label)) * 1/(sigma**2)))
    
    @property
    def adjacency_matrix(self):
        return self._adjacency_matrix * self.edge_weight
    
    @property
    def degree_matrix(self):
        return self._degree_matrix * self.edge_weight
    
    def observation_vector(self, positive=[], negative = []):
        """
        Take labels as observations being positive or negative
        and find the indices associated, along with a vector
        for those observations in the tree space
        
        Inputs
        ------
        
        positive list: list of labels that are positive
        negative list: list of labels that are negative
        
        Outputs
        -------
        tuple : positive feature indices, then negative feature
                indices as single list, along with an observation
                vector with +0.5 for positive and -0.5 for negative
                in the correct order for the observation.
        
        """
        pos_indices = [self.label2index[label] for label in positive]
        neg_indices = [self.label2index[label] for label in negative]
        return (pos_indices + neg_indices, np.concatenate([
            np.ones(len(pos_indices)) * 0.5, 
            np.ones(len(neg_indices)) * -0.5]))
    
    def _calculate_conditional_probabilities(self, y, indices, remaining_indices, cov):
        
        assert len(y) == len(indices), "yes"
        probs = np.zeros([len(remaining_indices), 2])
        for k, unseen_index in enumerate(remaining_indices):
            probs[k, 0] = norm_pdf(np.concatenate([y, [0.5]]),
                     cov[indices + [unseen_index], :][:, indices + [unseen_index]])
            probs[k, 1] = norm_pdf(np.concatenate([y, [-0.5]]),
                     cov[indices + [unseen_index], :][:, indices + [unseen_index]])
        # marginalize probabilities
        return (probs.T / probs.sum(axis=1)).T
    
    def conditional_probabilities(self,
                                  positive = [], 
                                  negative = [], 
                                  sigma = None,
                                  num_steps = 50,
                                  num_trials = 100,
                                  approximate = True):
        """
        Take labels as observations being positive or negative
        and find the conditional probability for each
        unassigned variable in the tree given the covariance matrix.
        
        Inputs
        ------
        
        positive list: list of labels that are positive
        negative list: list of labels that are negative
        sigma float : a prior variance on tree vertices
        num_steps int : how many steps in random walk for
                        Monte Carlo covariance technique
        num_trials int : how many samples of random walks to
                         use for each dimension in covariance
        approximate boolean : whether to use matrix inversion
                              or Monte Carlo to get covariance
                              matrix.
                              
        Outputs
        -------
        
        tuple <list<str> labels, np.array[n,2] probs> : return
            the labels of unassigned variables and their
            conditional probabilities
            
        """
        indices, y = self.observation_vector(positive,negative)
        Cov = self.approximate_covariance(
                sigma = sigma,
                num_steps = num_steps,
                num_trials = num_trials) if approximate else self.covariance(sigma)
        remaining_indices = list(
            set(range(0, len(self.index2label))) - set(indices)
        )
        
        probs = self._calculate_conditional_probabilities(y, indices, remaining_indices, Cov)
        
        return ([self.index2label[k] for k in remaining_indices], probs)
    
    def graph_conditional_probabilities(self,
                                        cmap = DEFAULT_COLORMAP,
                                        positive = [],
                                        negative = [],
                                        **kwargs):
        """
        Take labels as observations being positive or negative
        and find the conditional probability for each
        unassigned variable in the tree given the covariance matrix,
        and display it on a graph.
        
        Inputs
        ------
        
        positive list: list of labels that are positive
        negative list: list of labels that are negative
        sigma float : a prior variance on tree vertices
        num_steps int : how many steps in random walk for
                        Monte Carlo covariance technique
        num_trials int : how many samples of random walks to
                         use for each dimension in covariance
        approximate boolean : whether to use matrix inversion
                              or Monte Carlo to get covariance
                              matrix.
            
        """
        
        prob_graph = DrawableGraph(self._underlying_graph, cmap = cmap)
        
        labels, probs = self.conditional_probabilities(positive, negative,
                                                       **kwargs)
        prob_graph.apply_conditional_probabilities(labels, probs)
        prob_graph.apply_observations(positive, negative)
        return prob_graph
    
    def posterior_entropy(self,
                                  positive = [], 
                                  negative = [], 
                                  sigma = None,
                                  num_steps = 50,
                                  num_trials = 100,
                                  approximate = True):
        
        indices, y = self.observation_vector(positive,negative)
        Cov = self.approximate_covariance(
                sigma = sigma,
                num_steps = num_steps,
                num_trials = num_trials) if approximate else self.covariance(sigma)

        remaining_indices_set = set(range(0, len(self.index2label))) - set(indices)
        remaining_indices = list(remaining_indices_set)
        entropies = np.zeros([len(remaining_indices)])

        for k, unseen_index in enumerate(remaining_indices):
            # suppose we observe this element. What is the new entropy ?
            
            test_indices = indices + [unseen_index]
            
            prob_pos = norm_pdf(np.concatenate([y, [0.5]]),
                     Cov[test_indices, :][:, test_indices])
            prob_neg = norm_pdf(np.concatenate([y, [-0.5]]),
                     Cov[test_indices, :][:, test_indices])
            
            # marginalize:
            Z = (prob_pos + prob_neg)
            prob_pos /= Z
            prob_neg /= Z
            
            # get the entropy of remaining elements:
            post_indices = remaining_indices_set.copy()
            post_indices.remove(unseen_index)
            post_probs_pos = self._calculate_conditional_probabilities(np.concatenate([y , [0.5]]),
                                                                       test_indices,
                                                                       post_indices,
                                                                       Cov)
            post_probs_neg = self._calculate_conditional_probabilities(np.concatenate([y , [-.5]]),
                                                                       test_indices,
                                                                       post_indices,
                                                                       Cov)
            # expected entropy:
            entropies[k] = (
                prob_pos * entropy(post_probs_pos) +
                prob_neg * entropy(post_probs_neg)
            )
        return ([self.index2label[k] for k in remaining_indices], entropies)
        
        
```

```python
def display_cov_matrix(matrix, axis, labels = None, fontsize = 5):
    matshow = axis.matshow(matrix);
    if labels is not None:
        axis.set_yticks(np.arange(0, len(labels)))
        axis.set_yticklabels(labels, fontsize=fontsize);
        axis.set_xticks(np.arange(0, len(labels)))
        axis.set_xticklabels(labels, rotation = 'vertical', fontsize=fontsize);
    else:
        axis.get_xaxis().set_visible(False)
        axis.get_yaxis().set_visible(False)
    return matshow
    

def draw_covariance_matrices(graph, num_figs = 9, edge_range = [0.01, 8.0], normalize = False):
    fig = plt.figure()
    width_figs = int(np.sqrt(num_figs))
    height_figs = int(num_figs / width_figs)

    edge_values = np.linspace(edge_range[0], edge_range[1], num_figs)
    graph = DictTree(graph)

    for i in range(0, num_figs):
        graph.edge_weight = edge_values[i]

        # get graph stats:
        sigma = 5 # as in the paper we use 5
        laplacian = graph.laplacian()
        inv_cov = laplacian + np.eye(laplacian.shape[0]) * 1./(sigma**2)
        cov = np.linalg.inv(inv_cov)

        # use those to get a covariance matrix:
        ax = plt.subplot2grid((width_figs,height_figs), (int(i / height_figs), i % width_figs))
        # normalization
        cov = cov if normalize == False else cov / cov.max()
        if i == 0:
            matshow = display_cov_matrix( cov, ax, labels = graph.index2label)
        else:
            matshow = display_cov_matrix( cov, ax)
        cbar = fig.colorbar(matshow);
        cbar.ax.tick_params(labelsize=8)
        ax.text(0.5, -0.1,
                 "edge dist %.2f" % (edge_values[i]),
                 horizontalalignment='center',
                 transform = ax.transAxes,
                 fontsize=5);

    fig.text(0.5, 1.1, "Covariance Matrix over the tree",
                horizontalalignment='center', fontsize=15);
```

In our tree model with taxonomies we consider the probability of a feature to follow a multivariate gaussian distribution guided by the *covariance* of our tree:
$$\mathbb P(y | \mathrm{tree}) \sim \mathcal{N}(0, K),$$

where the covariance $K$ is given by [1, 2]:

$$K = \left(\Delta + \frac{1}{\sigma^2}\mathbf{I} \right)^{-1},$$

with $\Delta$ the laplacian of our Tree. Note how this is the formula for maximum likelihood estimation with prior information on $y$ given by $\frac{1}{\sigma^2}\mathbf{I}$. As in [2] we set the hyperparameter $\sigma = 5$.

1. Zhu, X., Lafferty, J., & Ghahramani, Z. (2003). *Semi-supervised learning: From Gaussian fields to Gaussian processes*  (Tech. Rep. No. CMU-CS- 03–175). Carnegie-Mellon University.

2. Kemp, C., Tenenbaum, J. (2009). *Structured Statistical Models of Inductive Reasoning* Psychological Review 2009, Vol. 116, No. 1, 20–58.

If we consider normalizing the covariance matrix $\Sigma_{\mathrm{normalized}} = \Sigma \cdot \max(\Sigma)^{-1}$, then varying the edge distance smoothly transitions the system from having strong connections all over (large edge distance) to functionning in isolation (small edge distance).

In this model $s_{i,j}$ thus corresponds to how strongly two pairwise elements are correlated (see low point of colormap above for intuition).

# A Monte Carlo approach

Let us consider the edge distance on the tree $s_{i,j}$ to be proportional to the likelihood of jumping from i to j. Then we can model this process as coin flipping with biased coins:
$$\mathbb p(i \to j) = s_{i,j},$$

If we also consider the variance term to be the likelihood of staying in place (self varying ?) then we can model this as:

$$\mathbb p(i \to j | i \to i = 0) = s_{i,j},$$

and

$$\mathbb p(i \to j | i \to i = 1) = \begin{cases} 1 & j = i,\\ 0 & \text{otherwise}\end{cases},$$

and $\mathbb p(i \to i) = \sigma^2$.

With no self connections ($\sigma^2 = 0$ essentially) we get:

```python
def plot_monte_carlo_cov(graph,
                          sigma = None,
                          edge_weight = 1.0,
                          num_steps = 50,
                          num_trials = 100,
                          fontsize = 10):
      graph = DictTree(graph, edge_weight = edge_weight)
      cov = graph.approximate_covariance(sigma = sigma,
                                         num_steps = num_steps,
                                         num_trials = num_trials)
      fig = plt.figure()
      ax = fig.add_axes([1, 1, 1, 1])
      matshow = display_cov_matrix(cov, ax, graph.index2label, fontsize=fontsize);
      cbar = fig.colorbar(matshow);
      cbar.ax.tick_params(labelsize=fontsize)
```

```python
plot_monte_carlo_cov(example_graph,
                          edge_weight= 0.1,
                          num_trials = 500)
```

When the self connection is high then we get higher baseline variance, and very low mobility (off diagonal terms are close to 0):

```python
plot_monte_carlo_cov(example_graph,
                          edge_weight= 0.1,
                          num_trials = 500,
                          sigma = 0.9)
```

If we gradually increase the mobility we can recover an intermediary state that appears proportional to the covariance matrix we originally wanted through a matrix inversion with $O( n \cdot t \cdot s)$ when $t$ is the number of trials, $s$ the number of steps, and $n$ the number of nodes in the tree (thus we recover an $O(n)$ algorithm, albeit a bad one) vs. our original $O(n^3)$ matrix inversion algorithm.

```python
plot_monte_carlo_cov(example_graph,
                          edge_weight= 0.1,
                          num_trials = 500,
                          sigma = 0.3)
```

```python
def show_conditional_probabilities(graph,
                                         positive=[],
                                         negative = [],
                                         sigma = None,
                                         edge_weight = 1.0,
                                         num_steps = 50,
                                         num_trials = 100):
    graph = DictTree(graph, edge_weight=edge_weight)
    labels, probs_mc = graph.conditional_probabilities(positive,
                                                    negative,
                                                    sigma,
                                                    num_steps,
                                                    num_trials,
                                                    approximate = True)
    labels, probs_inv = graph.conditional_probabilities(positive,
                                                    negative,
                                                    sigma,
                                                    approximate = False)
    fig = plt.figure()
    ax = fig.add_axes([1, 1, 1,1])
    ax.plot(probs_mc[:,0])
    ax.plot(probs_inv[:,0])
    ax.legend(["Approximate Covariance", "With Inversion"])
    ax.set_xticks(np.arange(0, len(labels)))
    ax.set_xticklabels(labels, fontsize = 8)
```

The approximate covariance doesnt get the same parent conditional likelihood due to too high mobility -- this is a hyperparameter than can be varied:

# Active probing

It is also of interest to ask which observation most reduces the posterior 

We may also consider the question of the lowest expected posterior Shannon entropy as in [1].

In the MC method the top most parent is not the probing with the highest reduction in entropy. Rather an intermediate location is preferred (**Politics**, **Religion**, or **Science**), and in the inversion case (could be considered more exact, but it in fact represents a slightly different mobility / covariance bias) the lowest expected posterior entropy is given by probing the topmost vertex.

[1] Johan De Kleer and Brian C. Williams, *Diagnosing Multiple Faults*, Artificial Intelligence, 
Volume 32, Issue 1, April 1987, pp 97 - 130

```python
mytree = DictTree(example_graph, edge_weight=0.1)
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
entropy_labels, post_entropies = mytree.posterior_entropy(["World news", "Tech news"],
sigma = 0.9,
num_steps = 10,
num_trials = 500,
approximate = True)
entropy_labels, post_entropies_exact = mytree.posterior_entropy(["World news", "Tech news"],
sigma = 0.9,
approximate = False)
ax.set_xticks(np.arange(len(entropy_labels)))
ax.set_xticklabels(entropy_labels, fontsize=8)
ax.plot(post_entropies);
ax.plot(post_entropies_exact);
ax.legend(["MC expected entropy", "Inversion expected entropy"]);
```

Suppose this observation is marked as true, then the posterior landscape for the inversion case is flat (as in the entropy of the posterior entropy is high), while the MC case still recommends probing middle-ground points.

```python
mytree = DictTree(example_graph, edge_weight=0.1)
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
entropy_labels, post_entropies = mytree.posterior_entropy(["World news", "Tech news", "World"],
sigma = 0.9,
num_steps = 10,
num_trials = 500,
approximate = True)
entropy_labels, post_entropies_exact = mytree.posterior_entropy(["World news", "Tech news", "World"],
sigma = 0.9,
approximate = False)
ax.set_xticks(np.arange(len(entropy_labels)))
ax.set_xticklabels(entropy_labels, fontsize=8)
ax.plot(post_entropies);
ax.plot(post_entropies_exact);
ax.legend(["MC expected entropy", "Inversion expected entropy"]);
```
