# Covariance over a tree

Get the covariance by using the Laplacian

```python
%load_ext autoreload
%autoreload 2
%config InlineBackend.figure_format = 'svg'
%matplotlib inline
import matplotlib.pyplot as plt
from IPython.display import Image
from graphviz import Digraph
import numpy as np

example_graph = {
    "World": {
    "News": {
        "World news": {},
        "Tech news": {}
    },
    "Science" : {
        "solar": {},
        "CS":{}
    },
    "Religion" : {
        "Temples":{},
        "Study":{}
    },
    "Politics": {
        "USA Politics": {},
        "French Politics": {}
    }
    }}
```

```python
def update_graph(graph, dict_node, code = [], name = None):
    if name == None and len(dict_node.keys()) > 1:
        name = 'ROOT'
    elif name == None and len(dict_node.keys()) == 1:
        return update_graph(graph, dict_node[list(dict_node.keys())[0]], code, list(dict_node.keys())[0])
    
    graph.node(str(code), "%s %r" % (name, code))
    # connect to parents
    if len(code) > 0:
        graph.edge(str(code[:-1]), str(code))
    
    for i, subkey in enumerate(dict_node.keys()):
        update_graph(graph, dict_node[subkey], code + [i], subkey)
    return graph

def convert_dict_to_matrix_labels(graph,
                                  entry_key = None,
                                  index2label = [],
                                  label2index = {},
                                  connections = {}):
    
    # create a root key if need be
    if entry_key == None and len(graph.keys()) == 1:
        entry_key = list(graph.keys())[0]
        graph = graph[entry_key]
    elif entry_key == None:
        entry_key = "ROOT"
    
    if label2index.get(entry_key, None) is not None:
        entry_key_index = label2index[entry_key]
    else:
        entry_key_index = len(label2index)
        label2index[entry_key] = entry_key_index
        index2label.append(entry_key)
    
    for subkey in graph.keys():
        # create child
        index2label.append(subkey)
        index = len(label2index)
        label2index[subkey] = index
        # connect parent to child:
        if connections.get(entry_key_index, None) is not None:
            connections[entry_key_index].append(index)
        else:
            connections[entry_key_index] = [index]
        
        # recurse:
        convert_dict_to_matrix_labels(graph[subkey],
                               entry_key = subkey,
                               connections = connections,
                               index2label = index2label,
                               label2index = label2index)
    return (index2label, label2index, connections)

def convert_dict_to_adjacency_matrix(graph):
    index2label, label2index, connections = convert_dict_to_matrix_labels(graph)
    adjacency_matrix = np.zeros([len(index2label),len(index2label)], dtype=np.int32)
    degree_matrix = np.zeros(len(index2label), dtype=np.int32)
    for key, connected_to in connections.items():
        adjacency_matrix[key, connected_to] = 1
        adjacency_matrix[connected_to, key] = 1
        degree_matrix[key] += len(connected_to)
        degree_matrix[connected_to] += 1
    return (index2label, label2index, connections, adjacency_matrix, degree_matrix)
```

```python
graph = Digraph(comment='The Round Table', format='png')
update_graph(graph, example_graph)
Image(graph.render('test-output/round-table', view=True))
```

```python
index2label, label2index, connections, adj_matrix, deg_matrix = convert_dict_to_adjacency_matrix(example_graph)
```

```python
# get graph stats:
sigma = 5 # as in the paper we use 5
laplacian = np.diag(deg_matrix) - adj_matrix
inv_cov = laplacian + np.eye(len(deg_matrix)) * 1./(sigma**2)
cov = np.linalg.inv(inv_cov)

# use those to get a covariance matrix:
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
matshow = ax.matshow(cov);
ax.set_yticks(np.arange(0, len(index2label)))
ax.set_yticklabels(index2label);
ax.set_xticks(np.arange(0, len(index2label)))
ax.set_xticklabels(index2label, rotation = 'vertical');
ax.text(0.5, 1.3, "Covariance Matrix over the tree",
         horizontalalignment='center', fontsize=15, transform = ax.transAxes)
fig.colorbar(matshow);
```
