# Intra-human agreement

We ask human subjects from Mechanical Turk to recommend a restaurant to a friend by looking at 3 places this *friend* has enjoyed in the past, and to select a new one from 3 others and compare with our recommender system's recommendations.

To empirically test whether ObjectLM along with our recommendation system correlates with human judgment we display recommendation scenarios to our system and human subjects. Each scenario is constructed by simulating 3 preferred locations for a fictional person, and 3 candidate places to recommend. Each fictional personal has a simple rule concering rating, categories, price, or value (difference between rating and price) that are used to sample from our database of restaurants this person's set of favorites and candidates. Each set of 3 candidates is built by taking 2 restaurants that do not follow the fictional person's selection rules, and 1 that does. Each set thus has an "expected winner," along with 2 "decoys". The workers on Mechanical Turk are then tasked with deciding which of the 3 candidates fits best given the 3 favorites.

A good source for Mechanical Turk api code can be [found here](http://www.toforge.com/2011/05/boto-mturk-tutorial-fetch-results-and-pay-workers/).

## Analyzing the results

The results from Mechanical Turk are useful for making 2 comparisions. Intra human worker agreement, and agreement between humans and recommendation system. The first metric allows us to evaluate the difficulty or existence of an actual ground truth. The second metric is the key result of our recommendation system that closely aligns with human judgments on this one-shot learning task. 


```python
def get_p_e():
    p_e = 0.
    different_cases = 0
    # probability of agreement, 5 people, 3 options:
    cases = [
        ([5, 0, 0], 3),
        ([4, 1, 0], 3 * 5 * 2),
        ([3, 2, 0], 3 * (5 * 4 / 2) * 2),
        ([3, 1, 1], 5 * 3 * 4 * 2)]

    for case, occ in cases:
        p_e += ((sum(i**2 for i in case) - sum(case)) / (sum(case) * (sum(case) - 1))) * occ
        different_cases += occ

    p_e /= different_cases
    return p_e

p_e = get_p_e()

def get_fleiss_kappa_scott(mtc, ex):
    results = [response for response in ex.get_responses(mtc) if (response['RejectionTime'] == None and response['ApprovalTime'] != None)]
    recommendations = Counter(result["recommendation"] for result in results)
    
    if len(recommendations.keys() - ex.option_names()) > 0:
        recommendations = Counter(result["recommendation"].split(" ")[0] for result in results)
        
    ratings = []
    for k, value in enumerate(recommendations.values()):
        ratings.extend([k] * value)
    return fleiss_kappa_single_item(ratings, 5, p_e)


def agreement(mtc, ex):
    results = [response for response in ex.get_responses(mtc) if (response['RejectionTime'] == None and response['ApprovalTime'] != None)]
    recommendations = Counter(result["recommendation"] for result in results)
    
    if len(recommendations.keys() - ex.option_names()) > 0:
        recommendations = Counter(result["recommendation"].split(" ")[0] for result in results)
    
    num_responses = sum(reco[1] for reco in recommendations.most_common())
    return recommendations.most_common(1)[0][1]

def distance_to_next_most_popular(mtc, ex):
    results = [response for response in ex.get_responses(mtc) if (response['RejectionTime'] == None and response['ApprovalTime'] != None)]
    recommendations = Counter(result["recommendation"] for result in results)
    
    if len(recommendations.keys() - ex.option_names()) > 0:
        recommendations = Counter(result["recommendation"].split(" ")[0] for result in results)
    
    if len(recommendations) > 1:
        return recommendations.most_common(2)[0][1] - recommendations.most_common(2)[1][1]
    else:
        return recommendations.most_common(1)[0][1]
```

### Mechanical Turk Polling

Simple utilities for getting mechanical turk results and approving them

```python
from personality import get_all_reviewable_hits

def approve_hits(mtc, min_answer_length = 2):
    reviewable_hits = get_all_reviewable_hits(mtc, verbose=False)
    unapproved = []
    for hit in reviewable_hits:
        assignments = mtc.get_assignments(hit.HITId)
        for assignment in assignments:
            if hasattr(assignment, 'ApprovalTime') or hasattr(assignment, 'RejectionTime'):
                continue
            answer_len = 0
            for answer in assignment.answers:
                for question_form_answer in answer:
                    if question_form_answer.qid == "explanation_comments":
                        answer_len= len(question_form_answer.fields[0].split(" "))
                        break
            if answer_len <= min_answer_length:
                unapproved.append((hit.HITId, assignment.AssignmentId))
                print("unapproved")
            else:
                mtc.approve_assignment(assignment.AssignmentId)
    return unapproved

def get_new_unapproved_assignments(mtc):
    reviewable_hits = get_all_reviewable_hits(mtc, verbose=False)
    found_guys = []
    for hit in reviewable_hits:
        assignments = mtc.get_assignments(hit.HITId)
        for assignment in assignments:
            if hasattr(assignment, 'ApprovalTime') or hasattr(assignment, 'RejectionTime'):
                continue
            else:
                found_guys.append((assignment, hit.HITId))
    return found_guys
```

```python
def fleiss_kappa(ratings, n):
    '''
    Computes the Fleiss' kappa measure for assessing the reliability of 
    agreement between a fixed number n of raters when assigning categorical
    ratings to a number of items.
    
    Args:
        ratings: a list of (item, category)-ratings
        n: number of raters
        k: number of categories
    Returns:
        the Fleiss' kappa score
    
    See also:
        http://en.wikipedia.org/wiki/Fleiss'_kappa
    '''
    items = set()
    categories = set()
    n_ij = {}
    
    for i, c in ratings:
        items.add(i)
        categories.add(c)
        n_ij[(i,c)] = n_ij.get((i,c), 0) + 1
    
    N = len(items)
    
    p_j = {}
    for c in categories:
        p_j[c] = sum(n_ij.get((i,c), 0) for i in items) / (1.0*n*N)
    
    P_i = {}
    for i in items:
        P_i[i] = (sum(n_ij.get((i,c), 0)**2 for c in categories)-n) / (n*(n-1.0))
    
    P_bar = sum(P_i.values()) / (1.0*N)
    P_e_bar = sum(p_j[c]**2 for c in categories)
    kappa = (P_bar - P_e_bar) / (1 - P_e_bar)
    
    return kappa

def fleiss_kappa_single_item(ratings, n, p_e):
    '''
    Computes the Fleiss' kappa measure for assessing the reliability of 
    agreement between a fixed number n of raters when assigning categorical
    ratings to a number of items.
    
    Args:
        ratings: a list of (item, category)-ratings
        n: number of raters
        k: number of categories
    Returns:
        the Fleiss' kappa score
    
    See also:
        http://en.wikipedia.org/wiki/Fleiss'_kappa
    '''
    categories = set()
    n_ij = {}
    
    for c in ratings:
        categories.add(c)
        n_ij[c] = n_ij.get(c, 0) + 1
    
    p_j = {}
    for c in categories:
        p_j[c] = n_ij.get(c, 0) / (1. * n)
    
    P_bar = (sum(n_ij.get(c, 0)**2 for c in categories)-n) / (n*(n-1.0))
    kappa = (P_bar - p_e) / (1 - p_e)
    
    return kappa
 
 
 
example = (                                                     [( 1,5)] * 14 +
                 [( 2,2)] * 2 + [( 2,3)] * 6  + [( 2,4)] * 4  + [( 2,5)] * 2  + 
                                [( 3,3)] * 3  + [( 3,4)] * 5  + [( 3,5)] * 6  + 
                 [( 4,2)] * 3 + [( 4,3)] * 9  + [( 4,4)] * 2  +
  [( 5,1)] * 2 + [( 5,2)] * 2 + [( 5,3)] * 8  + [( 5,4)] * 1  + [( 5,5)] * 1  + 
  [( 6,1)] * 7 + [( 6,2)] * 7 + 
  [( 7,1)] * 3 + [( 7,2)] * 2 + [( 7,3)] * 6  + [( 7,4)] * 3  + 
  [( 8,1)] * 2 + [( 8,2)] * 5 + [( 8,3)] * 3  + [( 8,4)] * 2  + [( 8,5)] * 2  + 
  [( 9,1)] * 6 + [( 9,2)] * 5 + [( 9,3)] * 2  + [( 9,4)] * 1  + 
                 [(10,2)] * 2 + [(10,3)] * 2  + [(10,4)] * 3  + [(10,5)] * 7  )
 
 
print('%.03f' % fleiss_kappa(example, 14)) # 0.210
```
